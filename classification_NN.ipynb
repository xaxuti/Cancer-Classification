```python
import tensorflow as tf
```

```python
from tensorflow.keras.models import Sequential
```

```python
from tensorflow.keras.layers import Dense
```

```python
data["diagnosis"].replace("M",1,inplace = True)
data["diagnosis"].replace("B",0,inplace = True)
```

```python
data.head()
```

```python
Y = data["diagnosis"]
```

```python
Y
```

```python
data.drop(["id","Unnamed: 32"],axis =1)
```

```python
Y
```

```python
X = data.drop(["id","Unnamed: 32","diagnosis"],axis =1)
```

```python
X
```

```python
norm = tf.keras.layers.Normalization(axis = 1)
```

```python
model = Sequential([
    
    Dense(units = 25,activation = "sigmoid"),
    Dense(units = 10,activation = "sigmoid"),
    Dense(units = 1,activation = "sigmoid")
])
```

```python
model.compile(
    loss = tf.keras.losses.BinaryCrossentropy(from_logits = True),
    optimizer = tf.keras.optimizers.Adam(0.01)
)
```

```python
from sklearn.model_selection import train_test_split
```

```python
data.dtypes
```

```python
x_train, x_test, y_train, y_test = train_test_split(X.iloc[1:], Y.iloc[1:], test_size=0.3, random_state=100) 
```

```python
model.fit(x_train,y_train,epochs = 100)
```

```python
prediction = model.predict(x_test)
```

```python
prediction
```

```python
import numpy as np
```

```python
y_pred = []
for i in prediction:
    if i >= 0.5:
        y_pred.append(1)
    else:
        y_pred.append(0)
y_hat = np.array(y_pred)
```

```python
y_hat.shape
```

```python
from sklearn import metrics
```

```python
MSE = metrics.mean_squared_error(y_test,y_hat)
MSE
```
